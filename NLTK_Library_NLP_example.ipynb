# **NLTK Kütüphanesi ile NLP Uygulaması**

```python
# Gerekli kütüphaneleri yükleme
!pip install nltk

# NLTK kütüphanesini içe aktarma
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# NLTK'nin gerektirdiği kaynakları indirme
nltk.download('punkt')

# Örnek metin belgesi
metin_belgesi = """
Doğal dil işleme (NLP), bilgisayarların insan dilini anlamasına ve yorumlamasına olanak tanıyan bir yapay zeka dalıdır.
NLP, metin madenciliği, dil modelleme ve metin analizi gibi birçok alt disiplini içerir.
"""

# Metni kelimelere ayırma
kelimeler = word_tokenize(metin_belgesi)

# Frekans dağılımını oluşturma
frekans_dagilimi = FreqDist(kelimeler)

# En sık kullanılan kelimeleri yazdırma
print("En sık kullanılan kelimeler:")
for kelime, frekans in frekans_dagilimi.most_common(5):
    print(f"{kelime}: {frekans}")

# Grafiği çizdirme
frekans_dagilimi.plot(20)
```

 Natural Language Toolkit ya da kısaltmasıyla NLTK, Python programlama dilinde doğal dil işleme (NLP) uygulamaları geliştirmek için kullanılan bir kütüphanedir. NLTK, dil analizi, dil modelleme, metin madenciliği ve diğer birçok dil işleme görevi için kullanılan araçları içerir. İşte NLTK'nin bazı temel özellikleri:

**Tokenization (Kelime Ayıklama):** Metni kelimelere veya cümlelere ayırma işlemidir. NLTK, metni bu tür parçalara ayırmak için kullanılabilir.

```python
from nltk.tokenize import word_tokenize
metin = "NLTK kelime ayıklama örneği."
kelimeler = word_tokenize(metin)
print(kelimeler)
```

**Stopwords (Durak Kelimeler):** Dil analizi sırasında genellikle önemsiz olan, sıklıkla kullanılan kelimeleri temsil eden durak kelimeleri (stopwords) çıkarmak önemlidir. NLTK içerisinde bu kelimelerin bir listesi bulunabilir.

```python
import nltk

# İlk çalıştırdığınızda, bu satırı kullanarak NLTK'nin indirme penceresini açabilirsiniz.
# nltk.download()

# Ancak, yukarıdaki satırı kullanmak yerine doğrudan durak kelimeleri indirebilirsiniz.
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

metin = "NLTK durak kelime örneği. Bu bir örnektir."
kelimeler = word_tokenize(metin)

durak_kelimeler = set(stopwords.words('turkish'))
temiz_kelimeler = [kelime for kelime in kelimeler if kelime.lower() not in durak_kelimeler]
print(temiz_kelimeler)
```

**Frekans Dağılımı (FreqDist):** Bir metindeki kelimelerin frekanslarını analiz etmek için kullanılır.

```python
from nltk.probability import FreqDist

frekans_dagilimi = FreqDist(kelimeler)
print(frekans_dagilimi.most_common(5))
```

**Stemming ve Lemmatization:** Kelimelerin köklerini (stemming) veya temel formlarını (lemmatization) elde etmek, metin madenciliği ve benzeri görevlerde önemlidir. NLTK bu konuda çeşitli araçlar sunar.

```python
import nltk

# İlk çalıştırdığınızda, bu satırı kullanarak NLTK'nin indirme penceresini açabilirsiniz.
# nltk.download()

# Ancak, yukarıdaki satırı kullanmak yerine doğrudan WordNet kaynağını indirebilirsiniz.
nltk.download('wordnet')
from nltk.stem import PorterStemmer, WordNetLemmatizer

porter_stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_word = porter_stemmer.stem("running")
lemmatized_word = lemmatizer.lemmatize("running")
print(f"Stemmed Word: {stemmed_word}")
print(f"Lemmatized Word: {lemmatized_word}")
```

**POS Tagging (Part-of-Speech Tagging):** Kelimelerin dilbilgisel özelliklerini belirleme işlemidir.

```python
import nltk

# İlk çalıştırdığınızda, bu satırı kullanarak NLTK'nin indirme penceresini açabilirsiniz.
# nltk.download()

# Ancak, yukarıdaki satırı kullanmak yerine doğrudan averaged_perceptron_tagger kaynağını indirebilirsiniz.
nltk.download('averaged_perceptron_tagger')

from nltk import pos_tag
from nltk.tokenize import word_tokenize

metin = "NLTK POS tagging örneği."
kelimeler = word_tokenize(metin)
tagged_words = pos_tag(kelimeler)
print(tagged_words)
```
